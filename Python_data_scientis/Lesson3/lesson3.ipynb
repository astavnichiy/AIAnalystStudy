```python
#3 Напишите код на Python, реализующий построение графиков:
#эллипс и круг

import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse



circle = plt.Circle((0.6, 0.8), 0.2, color='blue')
ellipse = Ellipse(xy=(0.2, 0.5), width=0.4, height=0.112, 
                        edgecolor='r', fc='None', lw=2)

fig, ax = plt.subplots() 

ax.add_artist(ellipse)
ax.add_artist(circle)

```

```python
#5. Задание (в программе)
#Нарисуйте трехмерный график двух параллельных плоскостей.

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
x = np.linspace(-1,1,10)
y = np.linspace(-1,1,10)

X,Y = np.meshgrid(x,y)

Z1=0.128*X + 0.001*Y + 0
Z2=0.128*X + 0.001*Y + 5

fig = plt.figure()
ax = fig.gca(projection='3d')
surf1 = ax.plot_surface(X, Y, Z1)
surf2 = ax.plot_surface(X, Y, Z2)


```
